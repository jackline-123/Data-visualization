# b)Bar graph

```python
titanic.Sex.value_counts(normalize=True).plot.bar(title='Gender Distribution',color=['gold','m'])
titanic.Sex.value_counts(normalize=True).round(3)*100
```

# Q2.Using Pandas visualization plot the bar the graph of the following features and add colors and report your results (You are advised to use percentage to show frequency of various values) (a) Pclass, b)Embarked, c)Parch¶

```python
#Pclass
titanic.Pclass.value_counts(normalize=True).plot.bar(title='Passenger Class Distribution',color=['m','r','b'])
titanic.Pclass.value_counts(normalize=True).round(3)*100
```

```python
#Embarked
titanic.Embarked.value_counts(normalize=True).plot.bar(title='Embarked Distribution',color=['m','y','g'])
titanic.Embarked.value_counts(normalize=True).round(3)*100
```

# There were many passengers who were embarked at S having a percentage of 72.4,18.9% of the passengers were embarked on C while 8.7% of the passengers embarked on Q.Q had the least passengers.

```python
#Parch
titanic.Parch.value_counts(normalize=True).plot.bar(title='Parch Distribution',color=['m','r','b'])
titanic.Parch.value_counts(normalize=True).round(3)*100
```

# Majority of the passengers were in parch 0 having a percentage of 76.1,parch 1 had 13.2% of the passengers,parch 2 had 9.0% of the passengers,parch 3 and 5 had equal number of passengers with a percentage of 0.6 each.Parch 4 had 0.4% of the passengers while parch 6 had the least number of passengers having apercentage of 0.1.
